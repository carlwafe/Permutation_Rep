```python
def add_weight_root(t):
    with t.clone() as tc:
        tc.set_label((), (tc.label(())[0]+1,tc.label(())[1]))
        return tc
```

```python
def star_map(s,t):
    return s.graft_on_root(t)
```

```python
## returns t_1.t_2.....t_n
def dot_map(trees):
    return LabelledRootedTree(trees,label=(0,0))
```

```python
## adds leaf at path
def add_leaf(t,p):
    with t.clone() as tc:
        tc.set_label(p, (t.label(p)[0],t.label(p)[1]+1))
        return tc
```

```python
## returns phi_i(s,t)
def phi(i,s,t):
    if i == 1:
        return add_weight_root(star_map(s,t))
    elif i == 2:
        return dot_map([s,add_weight_root(t)])
    elif i == 3:
        return star_map(s,add_weight_root(t))
    elif i == 4:
        return LabelledRootedTree([add_weight_root(star_map(s,list(iter(t))[0]))],label=(0,1))
    elif i == 5:
        return dot_map([s,add_weight_root(star_map(list(iter(t))[0],list(iter(t))[1]))])
    elif i == 6:
        return star_map(s,add_weight_root(star_map(list(iter(t))[0],list(iter(t))[1])))
    elif i == 7:
        return add_weight_root(star_map(t,s))
    elif i == 8:
        return star_map(t,add_weight_root(s))
    elif i == 9:
        return add_weight_root(star_map(star_map(s,list(iter(t))[0]),list(iter(t))[1]))
    elif i == 10:
        if s == LabelledRootedTree([],label=(0,2)):
            t_1 = list(iter(t))[0]
            return add_leaf(add_leaf(add_weight_root(dot_map([t_1,t_1])),()),())
        elif s.label(()) == (0,1) and children(s,()) == 1:
            s_2 = list(iter(s))[0]
            t_1 = list(iter(t))[0]
            return add_leaf(add_weight_root(dot_map([s_2,t_1,t_1])),())
        elif s.label(()) == (0,0) and children(s,()) == 2:
            s_1 = list(iter(s))[0]
            s_2 = list(iter(s))[1]
            t_1 = list(iter(t))[0]
            return add_weight_root(dot_map([s_1,s_2,t_1,t_1]))
    elif i == 11:
        if s.label(()) == (0,1) and children(s,()) == 1:
            s_2 = list(iter(s))[0]
            return LabelledRootedTree([add_weight_root(star_map(star_map(s_2,s_2),s_2))],label=(0,1))
        elif s.label(()) == (0,0) and children(s,()) == 2:
            s_1 = list(iter(s))[0]
            s_2 = list(iter(s))[1]
            t_1 = list(iter(t))[0]
            return dot_map([s_1,add_weight_root(star_map(star_map(s_2,t_1),t_1))])
```
