```python
#### Term i of A34, if text == True, prints terms 
def A34_hand_terms(n,i,text):
    #### Term 1
    result = 0
    if i == 1:
        if text: print("Term 1: ",n)
        else: return s[n]
    #### Term 2
    elif i == 2:
        for a in range(3,n+1):
            if n-a >= 4:
                temp = U(n,a)
                if not text: result += temp
                elif not temp == 0: print("Term 2: ",a,n-a)
    #### Term 3
    elif i == 3:
        for a in range(6,n+1):
            if n-a >= 1:
                temp = U(n,a)
                if not text: result += temp
                elif not temp == 0: print("Term 3: ",a,n-a)
    #### Term 4
    elif i == 4:
        for a in range(3,n+1):
            for b in range(4,n+1):
                if n-a-b >= 1:
                    temp = U3(n,a,b)
                    if not text: result += temp
                    elif not temp == 0: print("Term 4: ",a,b,n-a-b)
    #### Term 5
    elif i == 5:
        for a in range(3,n+1):
            for b in range(a+1,n+1):
                if n-a-b >= 2:
                    temp = U3(n,a,b)
                    if not text: result += temp
                    elif not temp == 0: print("Term 5: ",a,b,n-a-b)
        for a in range(3,n+1):
            if n-2*a >= 2 and a != n/3:
                temp = U3_s2(n,a)
                if not text: result += temp
                elif not temp == 0: print("Term 5: s_2",a,a,n-2*a)
    #### Term 6
    elif i == 6:
        for a in range(3,n+1):
            for b in range(5,n+1):
                temp = U3(n,a,b)
                if not text: result += temp
                elif not temp == 0: print("Term 6: ",a,b,n-a-b)
    ## Term 7
    elif i == 7:
        for a in range(4,n+1):
            for b in range(a+1,n+1):
                temp = U3(n,a,b)
                if not text: result += temp
                elif not U3(n,a,b) == 0: print("Term 7: ",a,b,n-a-b)
        for a in range(4,n+1):
            temp = U3_s2(n,a)
            if not text: result += temp
            elif not temp == 0: print("Term 7: s_2",a,a,n-2*a)
    #### Term 8
    elif i == 8:
        for a in range(3,n+1):
            for b in range(a+1,n+1):
                for c in range(2,n+1):
                    if n-a-b-c >= 1:
                        temp = U4(n,a,b,c)
                        if not text: result += temp
                        elif not temp == 0: print("Term 8: ",a,b,c,n-a-b-c)
        for a in range(3,n+1):
            for c in range(2,n+1):
                if n-2*a-c >= 1 and not (n-2*a-c == 1 and c == a):
                    temp = U4_s2(n,a,c)
                    if not text: result += temp
                    elif not temp == 0: print("Term 8: s_2 ",a,a,c,n-a-a-c)
    #### Term 9
    elif i == 9:
        for a in range(3,n+1):
            for b in range(2,n+1):
                for c in range(4,n+1):
                    temp = U4(n,a,b,c)
                    if not text: result += temp
                    elif not temp == 0: print("Term 9: ",a,b,c,n-a-b-c)
    #### Term 10
    elif i == 10:
        for a in range(3,n+1):
            for b in range(3,n+1):
                for c in range(3,n+1):
                    if a <= b:
                        temp = U4(n,a,b,c)
                        if not text: result += temp
                        elif not temp == 0: print("Term 10: ",a,b,c,n-a-b-c)
    #### Term 11
    elif i == 11:
        for a in range(3,n+1):
            for b in range(a+1,n+1):
                for c in range(b+1,n+1):
                    temp = U4(n,a,b,c)
                    if not text: result += temp
                    elif not temp == 0: print("Term 11: ",a,b,c,n-a-b-c)
        for a in range(3,n+1):
            for b in range(3,n+1):
                if not a == b and n-2*a-b >= 2:
                    temp = U4_s2(n,a,b)
                    if not text: result += temp
                    elif not temp == 0: print("Term 11: s_2",a,a,b,n-2*a-b)
        for a in range(3,n+1):
            if n-3*a >= 0:
                temp = U4_s3(n,a)
                if not text: result += temp
                elif not temp == 0: print("Term 11: s_3",a,a,a,n-3*a)
        
    #### Term 12
    elif i == 12:
        for a in range(4,n+1):
            for b in range(3,n+1):
                for c in range(b+1,n+1):
                    temp = U4(n,a,b,c)
                    if not text: result += temp
                    elif not temp == 0: print("Term 12: ",a,b,c,n-a-b-c)
        for a in range(4,n+1):
            for b in range(3,n+1):
                temp = U4_s2(n,b,a)
                if not text: result += temp
                elif not temp == 0: print("Term 12: s_2 ",b,b,a,n-2*b-a)
    #### Term 13
    elif i == 13:
        for a in range(3,n+1):
            for b in range(a+1,n+1):
                for c in range(b+1,n+1):
                    temp = U5(n,a,b,c,1)
                    if not text: result += temp
                    elif not temp == 0: print("Term 13: ",a,b,c,1,n-a-b-c-1)
        for a in range(3,n+1):
            for b in range(3,n+1):
                if a != b:
                    temp = U5_s2(n,a,b,1)
                    if not text: result += temp
                    elif not temp == 0: print("Term 13: s_2",a,a,b,1,n-2*a-b-1)
        for a in range(3,n+1):
            temp = U5_s3(n,a,1)
            if not text: result += temp
            elif not temp == 0: print("Term 13: s_3",a,a,a,1,n-3*a-1)
    #### Term 14
    elif i == 14:
        for a in range(3,n+1):
            for b in range(2,n+1):
                for c in range(3,n+1):
                    for d in range(c+1,n+1):
                        temp = U5(n,a,b,c,d)
                        if not text: result += temp
                        elif not temp == 0: print("Term 14: ",a,b,c,d,n-a-b-c-d)
        for a in range(3,n+1):
            for b in range(2,n+1):
                for c in range(3,n+1):
                    if n-2*c-a-b >= 1:
                        temp = U5_s2(n,c,a,b)
                        if not text: result += temp
                        elif not temp == 0: print("Term 14: s_2",a,b,c,c,n-2*c-a-b)
    #### Term 15
    elif i == 15:
        # Case a < b
        for a in range(3,n+1):
            for b in range(a+1,n+1):
                for c in range(2,n+1):
                    for d in range(2,n+1):
                        temp = U5(n,a,b,c,d)
                        if not text: result += temp
                        elif not temp == 0: print("Term 15: ",a,b,c,d,n-a-b-c-d)
        # Case a = b and c < d
        for a in range(3,n+1):
            for c in range(2,n+1):
                    for d in range(c+1,n+1):
                        temp = U5(n,a,a,c,d)
                        if not text: result += temp
                        elif not temp == 0: print("Term 15: ",a,a,c,d,n-2*a-c-d)
        # Case a = b and c = d
        for a in range(3,n+1):
            for c in range(2,n+1):
                e = n-2*a-2*c
                if e >= 0:
                    temp = s[2].plethysm(U(a+c,a))*s[e]
                    if not text: result += temp
                    elif not temp == 0:
                        print("Term 15 (s2\circ U_%d,%d) * s([%d])" % (a,c,e))
    #### Term 16
    elif i == 16:
        for a in range(3,n+1):
            for b in range(a+1,n+1):
                for c in range(b+1,n+1):
                    for d in range(c+1,n+1):
                        temp = U5(n,a,b,c,d)
                        if not text: result += temp
                        elif not temp == 0: print("Term 16: ",a,b,c,d,n-a-b-c-d)
        for a in range(3,n+1):
            for b in range(3,n+1):
                for c in range(3,n+1):
                    if a != b and a != c:
                        temp = U5_s2(n,a,b,c)
                        if not text: result += temp
                        elif not temp == 0: print("Term 16: s_2 ",a,a,b,c,n-a-b-c)
        for a in range(3,n+1):
            for b in range(3,n+1):
                if a != b:
                    temp = U5_s3(n,a,b)
                    if not text: result += temp
                    elif not temp == 0: print("Term 16: s_3",a,a,a,b,n-a-b-c)
        for a in range(3,n+1):
            temp = U5_s4(n,a)
            if not text: result += temp
            elif not temp == 0: print("Term 16: s_4 ",a,a,a,a,n-4*a)
    if text == False:
        return result
```

```python
# If text == False returns A^3(M_0,n)+A^4(M_0,n) 
# If text == True shows which terms contribute to A^3(M_0,n)+A^4(M_0,n) 
def A34_hand(n,text):
    if text:
        for i in range(1,17):
            A34_hand_terms(n,i,text)
    else:
        result = 0
        for i in range(1,17):
            result += A34_hand_terms(n,i,text)
        return result
```
