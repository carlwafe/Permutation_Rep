```python
# If text == False returns A^4(M_0,n)
# If text == True shows which terms contribute to A^4(M_0,n) 
def A4_hand_terms(n,i,text):
    result = 0
    ## Term 1
    if i == 1:
        if n%2 == 1:
            result += s[n]
            if text == True: print("Term1: ",n)

    ## Term 2
    if i == 2:
        for a in range(3,n+1):
            b = n-a
            if b >= 4:
                result += U(n,a)
                if not U(n,a) == 0 and text == True: print("Term 2: ",a,b)

    ## Term 3
    if i == 3:
        for a in range(7,n+1):
            b = n-a
            if a%2 == 1 and b >= 1:
                result += U(n,a)
                if not U(n,a) == 0 and text == True: print("Term 3: ",a,b)

    ## Term 4
    if i == 4:
        for a in range(3,n+1):
            if n-a-1 >= 4:
                result += U3(n,a,1)
                if not U3(n,a,1) == 0 and text == True: print("Term 4: ",a,n-a-1,1)
        for a in range(int(np.ceil((n-2)/2)),n+1):
            if n-a-2 >= 4:
                result += U3(n,a,2)
                if not U3(n,a,2) == 0 and text == True: print("Term 4: ",a,n-a-2,2)
        for a in range(4,n+1):
            for b in range(a+1,n+1):
                if n-a-b >= 3:
                    result += U3(n,a,b)
                    if not U3(n,a,b) == 0 and text == True: print("Term 4: ",a,b,n-a-b)

    ## Term 5
    if i == 5:
        for a in range(5,n+1):
            for b in range(a+1,n+1):
                if n-a-b == 3 or n-a-b == 4:
                    temp = U3(n,a,b)
                    result += temp
                    if not temp == 0 and text == True: print("Term 5: ",a,b,n-a-b)
        temp = U3(n,3,n-5)
        result += temp
        if not temp == 0 and text == True: print("Term 5:",a,n-5,2)
        for a in range(4,n+1):
            for b in range(a+1,n+1):
                if n-a-b == 2 or n-a-b >= 5:
                    temp = U3(n,a,b)
                    result += temp
                    if not temp == 0 and text == True: print("Term 5: ",a,b,n-a-b)
        for a in range(3,n+1):
            if n-2*a >= 2 and not a == n/3:
                temp = U3_s2(n,a)
                result += temp
                if not temp == 0 and text == True: print("Term 5: s_2",a,a,n-2*a)
                
    ## Term 6 is empty
    
    ## Term 7
    if i == 7:
        for a in range(5,n+1):
            for b in range(a+1,n+1):
                temp = U3(n,a,b)
                result += temp
                if not temp == 0 and text == True: print("Term 6: ",a,b,n-a-b)
        for a in range(4,n+1):
            if not a == (n-2)/2:
                temp = U3_s2(n,a)
                result += temp
                if not temp == 0 and text == True: print("Term 6: s_2",a,a,n-2*a)
        if n%3 == 0:
            temp = U3_s2(n,n/3)
            result += temp
            if not temp == 0 and text == True: print("Term 6: s_2",n/3,n/3,n/3)

    # Term 8
    if i == 8:
        for a in range(3,n+1):
            for b in range(a+1,n+1):
                if b > (n-2)/2 and n-a-b-1 >= 2:
                    temp = U4(n,a,b,1)
                    result += temp
                    if not temp == 0 and text == True: print("Term 8: ",a,b,n-a-b-1,1)
        for a in range(3,n+1):
            for b in range(a+1,n+1):
                for c in range(2,n+1):
                    if n-a-b-c >= 2:
                        temp = U4(n,a,b,c)
                        result += temp
                        if not temp == 0 and text == True: print("Term 8: ",a,b,c,n-a-b-c)
        for a in range(3,n+1):
            for c in range(2,n+1):
                if n-2*a-c >= 2:
                    temp = U4_s2(n,a,c)
                    result += temp
                    if not temp == 0 and text == True: print("Term 8: s_2 ",a,a,c,n-a-a-c)
        if n == 10:
            temp = U4_s2(10,3,3)
            result -= temp
            if not temp == 0 and text == True: print("n=10 and we have -Term 8: s_2 ",3,3,3,1)
                
    ## Term 9
    if i == 9:
        for a in range(4,n+1):
            for c in range(4,n+1):
                for b in range(2,n+1):
                    if a+b > n/2:
                        result += U4(n,a,b,c)
                        if not U4(n,a,b,c) == 0 and text == True: print("Term 9: ",a,b,c,n-a-b-c)
        for a in range(3,n+1):
            for b in range(2,n+1):
                if n-a-b >= 4 and a+b < n/2 and not (a==3 and b==3):
                    result += U3(n,a,b)
                    if not U3(n,a,b) == 0 and text == True: print("Term 9: ",a,b,n-a-b)
        if n%2 == 0:
            for b in range(2,n/2-3+1):
                for d in range(0,n/2-4+1):
                    if b >= d:
                        result += U4(n,n/2-b,b,n/2-d)
                        if not U4(n,n/2-b,b,n/2-d) == 0 and text == True: print("Term 9: ",n/2-b,b,n/2-d,d)
            result -= U4(n,n/2-2,2,3)
            if not U4(n,n/2-2,2,3) == 0 and text == True: print("Term 9: -",n/2-2,2,3,n/2-3)
        for a in range(3,n+1):
            for b in range(2,n+1):
                if a+b < n/2:
                    result -= U4(n,3,3,a)
                    if not U4(n,3,3,a) == 0 and text == True: print("Term 9: -",a,b,3,3)

    ## Term 10
    if i == 10:
        for a in range(3,n+1):
            for b in range(a,n+1):
                for c in range(3,n+1):
                     result += U4(n,a,b,c)
                        
    ## Term 11 is empty
    
    ## Term 12
    if i == 12:
        for a in range(4,n+1):
            for b in range(3,n+1):
                for c in range(b+1,n+1):
                    result += U4(n,a,b,c)
                    if not U4(n,a,b,c) == 0 and text == True: print("Term 11: ",a,b,c,n-a-b-c)
        for a in range(4,n+1):
            for b in range(3,n+1):
                if not n-2*b-a == 1 and not b == (n-1)/3:
                    result += U4_s2(n,b,a)
                    if not U4_s2(n,b,a) == 0 and text == True: print("Term 11: s_2 ",b,b,a,n-2*b-a)

    ## Term 13
    if i == 13:
        for a in range(3,n+1):
            for b in range(a+1,n+1):
                for c in range(b+1,n+1):
                    result += U5(n,a,b,c,1)
                    if not U5(n,a,b,c,1) == 0 and text == True: print("Term 13: ",a,b,c,1,n-a-b-c-1)
        for a in range(3,n+1):
            for b in range(3,n+1):
                if a != b and n-2*a-b-1 >= 1:
                    result += U5_s2(n,a,b,1)
                    if not U5_s2(n,a,b,1) == 0 and text == True: print("Term 13: s_2",a,a,b,1,n-2*a-b-1)
        for a in range(3,n+1):
            result += U5_s3(n,a,1)
            if not U5_s3(n,a,1) == 0 and text == True: print("Term 13: s_3",a,a,a,1,n-3*a-1)

    ## Term 14
    if i == 14:
        for a in range(3,n+1):
            for b in range(2,n+1):
                for c in range(3,n+1):
                    for d in range(c+1,n+1):
                        if n-a-b-c-d >= 1:
                            result += U5(n,a,b,c,d)
                            if not U5(n,a,b,c,d) == 0 and text == True: print("Term 14: ",a,b,c,d,n-a-b-c-d)
        for a in range(3,n+1):
            for b in range(2,n+1):
                for c in range(3,n+1):
                    result += U5_s2(n,c,a,b)
                    if not U5_s2(n,c,a,b) == 0 and text == True: print("Term 14: s_2",c,c,a,b,n-2*c-a-b)
        for a in range(3,n+1):
            for b in range(2,n+1):
                for c in range(4,n+1):
                    if c < n-a-b-c:
                        result += U4(n,a,b,c)
                        if not U4(n,a,b,c) == 0 and text == True: print("Term 14: ",a,b,c,n-a-b-c)
        for a in range(3,n+1):
            for b in range(2,n+1):
                if a+b >= n/2 and n-a-b-3 >= 4 and not b == n/2-3:
                    result += U4(n,a,b,3)
                    if not U4(n,a,b,3) == 0 and text == True: print("Term 14: ",a,b,3,n-a-b-3)

    ## Term 15
    if i == 15:
        for a in range(3,n+1):
            for b in range(a+1,n+1):
                for c in range(2,n+1):
                    for d in range(c+1,n+1):
                        result += U5(n,a,b,c,d)
        ##############################################################
        ########## Term with S_2textS_2 symmetry still missing ##########
        ########## Need to define U4_s2(a,a,b,b,c) ###################
        ##############################################################
        if n == 11:
            result += s([2]).plethysm(U(5,3))*s([1])

    ## Term 16
    if i == 16:
        for a in range(3,n+1):
            for b in range(a+1,n+1):
                for c in range(b+1,n+1):
                    for d in range(c+1,n+1):
                        result += U5(n,a,b,c,d)
        for a in range(3,n+1):
            for b in range(3,n+1):
                for c in range(3,n+1):
                    if a != b and a != c:
                        result += U5_s2(n,a,b,c)
        for a in range(3,n+1):
            for b in range(3,n+1):
                if a != b:
                    result += U5_s3(n,a,b)
        for a in range(3,n+1):
            result += U5_s4(n,a)
        
    if text == False:
        return result
```

```python
def A4_hand(n,text):
    if text:
        for i in range(1,17):
            A4_hand_terms(n,i,text)
    else:
        result = 0
        for i in range(1,17):
            result += A4_hand_terms(n,i,text)
        return result
```

```python
# If x == False returns A^3(M_0,n) 
# If x == True shows which terms contribute to A^3(M_0,n) 
def A3_hand(n,x):
    result = 0
    ## Term 1
    for a in range(6,n+1):
        if a%2 == 0:
            result += U(n,a)
            if not U(n,a) == 0 and x == True: print("Term 1: ",a,n-a)
    
    ## Term 2
    for a in range(4,n+1):
        for b in range(3,n+1):
            result += U3(n,a,b)
            if not U3(n,a,b) == 0 and x == True: print("Term 2: ",a,b,n-a-b)
    
    ## Term 3
    for a in range(3,n+1):
        for b in range(a,n+1):
            if n-a-b >= 2 and not (a==(n-2)/2 and b==(n-2)/2 and n-a-b==2):
                result += U3(n,a,b)
                if not U3(n,a,b) == 0 and x == True: print("Term 3: ",a,b,n-a-b)
    
    ## Term 4
    for a in range(3,n+1):
        for b in range(a+1,n+1):
            if b <= (n-2)/2:
                result += U4(n,a,b,1)
                if not U4(n,a,b,1) == 0 and x == True: print("Term 4: ",a,b,1,n-a-b-1)
    for a in range(3,n+1):
        if a < (n-2)/2:
            result += U4_s2(n,a,1)
            if not U4_s2(n,a,1) == 0 and x == True: print("Term 4: s_2 ",a,a,1,n-2*a-1)
    
    ## Term 5
    for a in range(3,n+1):
        for b in range(2,n+1):
            for c in range(3,n+1):
                if n-a-b-c >= 1 and (a+b < n/2 or (a+b == n/2 and b < n-a-b-c)):
                    result += U4(n,a,b,c)
                    if not U4(n,a,b,c) == 0 and x == True: print("Term 5: ",a,b,c,n-a-b-c)
                                     
    ## Term 6
    for a in range(3,n+1):
        for b in range(a+1,n+1):
            for c in range(b+1,n+1):
                result += U4(n,a,b,c)
                if not U4(n,a,b,c) == 0 and x == True: print("Term 6: ",a,b,c,n-a-b-c)
    for a in range(3,n+1):
        for b in range(3,n+1):
            if (not a == b) and n-2*a-b >= 1:
                result += U4_s2(n,a,b)
                if not U4_s2(n,a,b) == 0 and x == True: print("Term 6: s_2 ",a,a,b,n-2*a-b)
    for a in range(3,n+1):
        result += U4_s3(n,a)
        if not U4_s3(n,a) == 0 and x == True: print("Term 6: s_3 ",a,a,a,n-3*a)
    
    ## Term 7
    if n%3 == 0:
        result -= U3_s2(n,n/3)
        if not U3_s2(n,n/3) == 0 and x == True: print("Term 7: -s_2 ",n/3,n/3,n/3)
            
    ## Term 8
    if n%2 == 0:
        result += U3_s2(n,(n-2)/2)
        if not U3_s2(n,(n-2)/2) == 0 and x == True: print("Term 8: s_2 ",(n-2)/2,(n-2)/2,2)
    
    ## Term 9
    if n%2 == 0:
        for a in range(3,n+1):
            if a <= n/2-2:
                result += s([2]).plethysm(U(n/2,a))
                if not U(n/2,a) == 0 and x == True: print("Term 9 plethysm term s_s\circ U_",a,n/2-a)
    
    if x == False:
        return result
```
