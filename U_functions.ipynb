```python
## Function returns U_{a,n-a}
def U(n,a):
    if n <= 0:
        return 0
    elif not (a == int(a) and n == int(n)):
        return 0
    elif a == 0 or a == n:
        return s[n]
    elif a > 0 and a < n:
        return s([n-a])*s([a])
    else: # this is called if a is an integer with a < 0 or a > n
        return 0
```

```python
## Function returns U_{a,b,n-a-b}
def U3(n,a,b):
    if n <= 0:
        return 0
    elif not (a == int(a) and b == int(b) and n == int(n)):
        return 0
    elif a == 0:
        return U(n,b)
    elif b == 0:
        return U(n,a)
    elif a + b == n:
        return U(n,a)
    elif b > 0 and a+b < n and n-a-b > 0:
        return U(a+b,a)*s([n-a-b])
    else:
        return 0
```

```python
def U3_s2(n,a):
    if n <= 0:
        return 0
    elif not a == int(a):
        return 0
    elif a == 0:
        return  s([n])
    elif 2*a == n:
        return s([2]).plethysm(s([a]))
    elif a > 0 and n-2*a > 0:
        return (s([2]).plethysm(s([a])))*s([n-2*a])
    else:
        return 0
def U3_s3(n):
    if n <= 0:
        return 0
    elif not (n == int(n) and n%3 == 0):
        return 0
    return s([3]).plethysm(s[n/3])
```

```python
def U4(n,a,b,c):
    if n <= 0:
        return 0
    elif not c == int(c):
        return 0
    if a == 0:
        return U3(n,b,c)
    elif b == 0:
        return U3(n,a,c)
    elif c == 0:
        return U3(n,a,b)
    elif a+b+c == n:
        return U3(n,a,b)
    elif a > 0 and b > 0 and n-a-b-c > 0: ## note that a < n, b < n and n-a-b-c < n are automatic
        return U3(a+b+c,a,b)*s([n-a-b-c]) ## also note a < n, b < n are not even necessary to check here, as they are dealt with in U3
    else:
        return 0
## Outputs U^{S_2}_{a,a,b,n-2a-b} 
def U4_s2(n,a,b):
    if n <= 0:
        return 0
    elif not (a == int(a) and b == int(b)):
        return 0
    elif n-2*a-b == 0:
        return U3_s2(n,a)
    elif a == 0:
        return U(n,b)
    elif b == 0:
        return U3_s2(n,a)
    elif a > 0 and b > 0 and n-2*a-b > 0: # note that a < n, b < n and n-2*a-b < n are automatic
        return U3_s2(2*a+b,a)*s([n-2*a-b])
    else:
        return 0
## Outputs U^{S_3}_{a,a,a,n-3*a}
def U4_s3(n,a):
    if n <= 0:
        return 0
    elif not (a == int(a) and n == int(n)):
        return 0
    elif a == 0:
        return s([n])
    elif n-3*a == 0:
        return U3_s3(n)
    elif a > 0 and n-3*a > 0: ## note that n-3*a < n and 3*a < n are automatic
        return (s([3]).plethysm(s([a])))*s([n-3*a])
    else:
        return 0
```

```python
## Output U_{a,b,c,d,n-a-b-c-d}
def U5(n,a,b,c,d):
    if n <= 0:
        return 0
    elif not d == int(d):
        return 0
    if a == 0:
        return U4(n,b,c,d)
    elif b == 0:
        return U4(n,a,c,d)
    elif c == 0:
        return U4(n,a,b,d)
    elif d == 0:
        return U4(n,a,c,b)
    elif a+b+c+d == n:
        return U4(n,a,b,c)
    elif a > 0 and b > 0 and c > 0 and n-a-b-c-d > 0: # note that a < n, b < n, c < n, d < n and n-a-b-c-d < n are automatic
        return U4(a+b+c+d,a,b,c)*s([n-a-b-c-d])
    else:
        return 0
## Outputs U^{S_2}_{a,a,b,c,n-2a-b-c} 
def U5_s2(n,a,b,c):
    if n <= 0:
        return 0
    elif not (a == int(a) and b == int(b) and c == int(c)):
        return 0
    elif a == 0:
        return U3(n,b,c)
    elif b == 0:
        return U4_s2(n,a,c)
    elif c == 0:
        return U4_s2(n,a,b)
    elif n-2*a-b-c == 0:
        return U4_s2(n,a,b)
    elif a > 0 and b > 0 and c > 0 and n-2*a-b-c > 0: ## note that 2*a+b+c < n is automatic
        return U4_s2(2*a+b+c,a,b)*s([n-2*a-b-c])
    else:
        return 0
## Outputs U^{S_3}_{a,a,a,b,n-3*a-b}
def U5_s3(n,a,b):
    if n <= 0:
        return 0
    elif not (a == int(a) and b == int(b) and n == int(n)):
        return 0
    elif a == 0:
        return U(n,b)
    elif b == 0 or n-3*a-b == 0:
        return U4_s3(n,a)
    elif a > 0 and b > 0  and n-3*a-b > 0:
        return (s([3]).plethysm(s([a])))*s([b])*s([n-3*a-b])
    else:
        return 0
## Outputs U^{S_4}_{a,a,a,a,n-4*a}
def U5_s4(n,a):
    if n <= 0:
        return 0
    elif not (a == int(a) and n == int(n)):
        return 0
    elif a > 0 and n-4*a >= 0:
        return (s([4]).plethysm(s([a])))*s([n-4*a])
    else:
        return 0
```
