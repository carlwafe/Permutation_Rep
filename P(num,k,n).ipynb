```python
## Given a tree t returns the bare tree corresponging to it
def bare_tree(t):
    with t.clone() as tc:
        for p in list(OrderedTree(tc).paths()):
            tc.set_label(p, (tc.label(p)[0],0))
        return tc
```

```python
zero_tree = LabelledRootedTree([],label=(0,0))
one_tree = LabelledRootedTree([],label=(1,0))
two_tree = LabelledRootedTree([],label=(2,0))
three_tree = LabelledRootedTree([],label=(3,0))
one_one_tree = LabelledRootedTree([one_tree],label=(1,0))
zero_one_tree = LabelledRootedTree([one_tree],label=(0,0))
```

```python
def order_trees(s,t):
    if total_weight(s) > total_weight(t):
        return False
    elif total_weight(s) == total_weight(t) and (total_weight(s) == 0 or total_weight(s) == 1):
        if bare_tree(s) == bare_tree(t):
            if bare_tree(s) == zero_tree or bare_tree(s) == one_tree:
                if s.label(())[1] <= t.label(())[1]:
                    return True
                else:
                    return False
            elif bare_tree(s) == zero_one_tree:
                if s.label(())[1] + s.label((0,))[1] < t.label(())[1] + t.label((0,))[1]:
                    return True
                elif number_of_leafs(s) == number_of_leafs(t):
                    if s.label(())[1] <=  t.label(())[1]:
                        return True
                    else:
                        return False
        elif bare_tree(s) == one_tree and bare_tree(t) == zero_one_tree:
            return True
        elif bare_tree(s) == zero_one_tree and bare_tree(t) == one_tree:
            return False
    elif total_weight(s) == total_weight(t):
        raise ValueError("order not defined for s,t")
    elif total_weight(s) < total_weight(t):
        return True
```

```python
##### Returns P_num for num = 1,...,11 for weight k and leafs n
def P(num,k,n):
    result = []
    for i in range(0,k): ### Could make faster for up to int((k-1)/2+1)????
            for h in range(2,(n+1)-2):
                for s in list_trees(i,h):
                    for t in list_trees(k-1-i,n-h):
                        if order_trees(s,t):
                            ##### P_1 #####
                            if num == 1:
                                if valency(t,()) >= 4:
                                    if t.label(())[0] > 0:
                                        result.append((s,t))
                            ##### P_2 #####
                            elif num == 2:
                                if valency(t,()) >= 4:
                                    if s.label(())[0] > 0 and t.label(())[0] == 0:
                                        result.append((s,t))
                            ##### P_3 #####
                            elif num == 3:
                                if valency(t,()) >= 4:
                                    if s.label(())[0] == 0 and t.label(())[0] == 0:
                                        result.append((s,t))
                            ##### P_4 #####
                            elif num == 4:
                                if not s == t and t.label(()) == (0,1) and children(t,()) == 1:
                                    result.append((s,t))
                            ##### P_5 #####
                            elif num == 5:
                                if t.label(()) == (0,0) and children(t,()) == 2:
                                    t_1 = list(iter(t))[0]
                                    t_2 = list(iter(t))[1]
                                    if order_trees(t_1,t_2) and not t_1 == t_2 and s.label(())[0] > 0:
                                        result.append((s,t))
                            ##### P_6 #####
                            elif num == 6:
                                if t.label(()) == (0,0) and children(t,()) == 2:
                                    t_1 = list(iter(t))[0]
                                    t_2 = list(iter(t))[1]
                                    if order_trees(t_1,t_2) and not t_1 == t_2 and s.label(())[0] == 0:
                                        result.append((s,t))
                            ##### P_7 #####
                            elif num == 7:
                                if valency(s,()) >= 4 and t.label(()) == (0,0) and children(t,()) == 2:
                                    t_1 = list(iter(t))[0]
                                    t_2 = list(iter(t))[1]
                                    if t_1 == t_2:
                                        if s.label(())[0] > 0 and not s == t_1:
                                            result.append((s,t))
                            ##### P_8 #####
                            elif num == 8:
                                if valency(s,()) >= 4 and t.label(()) == (0,0) and children(t,()) == 2:
                                    t_1 = list(iter(t))[0]
                                    t_2 = list(iter(t))[1]
                                    if t_1 == t_2:
                                        with s.clone() as sc:
                                            sc.set_label((), (sc.label(())[0]+1,sc.label(())[1]))
                                            if s.label(())[0] == 0 and not sc == t_1:
                                                result.append((s,t))
                            ##### P_9 #####
                            elif num == 9:
                                if valency(s,()) >= 4 and t.label(()) == (0,0) and children(t,()) == 2:
                                    t_1 = list(iter(t))[0]
                                    t_2 = list(iter(t))[1]
                                    if t_1 == t_2:
                                        if s == t_1:
                                            result.append((s,t))
                                        elif s.label(())[0] == 0:
                                            with s.clone() as sc:
                                                sc.set_label((), (sc.label(())[0]+1,sc.label(())[1]))
                                                if sc == t_1:
                                                    result.append((s,t))
                            ###### P_10 ######
                            elif num == 10:
                                if t.label(()) == (0,0) and children(t,()) == 2:
                                    t_1 = list(iter(t))[0]
                                    t_2 = list(iter(t))[1]
                                    if t_1 == t_2:
                                        if s == LabelledRootedTree([],label=(0,2)):
                                            result.append((s,t))
                                        elif s.label(()) == (0,1) and children(s,()) == 1:
                                            s_2 = list(iter(s))[0]
                                            if not s_2 == t_1:
                                                result.append((s,t))
                                        elif s.label(()) == (0,0) and children(s,()) == 2:
                                            s_1 = list(iter(s))[0]
                                            s_2 = list(iter(s))[1]
                                            if not s_2 == t_1:
                                                result.append((s,t))
                            ###### P_11 ######
                            elif num == 11:
                                if t.label(()) == (0,0) and children(t,()) == 2:
                                    t_1 = list(iter(t))[0]
                                    t_2 = list(iter(t))[1]
                                    if t_1 == t_2:
                                        if s.label(()) == (0,1) and children(s,()) == 1:
                                            s_2 = list(iter(s))[0]
                                            if s_2 == t_1:
                                                result.append((s,t))
                                        elif s.label(()) == (0,0) and children(s,()) == 2:
                                            s_1 = list(iter(s))[0]
                                            s_2 = list(iter(s))[1]
                                            if not s_1 == s_2 and s_2 == t_1:
                                                result.append((s,t))
    return result
```
